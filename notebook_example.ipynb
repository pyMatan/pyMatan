```python
from pyMatan.integrals import IntegralSolver
import numpy as np
import sympy as sp

print("--- ТЕСТ 1: x^2 ---")
solver = IntegralSolver("x**2")

res_trap = solver.trapezoidal(0, 1, 100)
res_simp = solver.simpson(0, 1, 100)

print(f"Трапеції: {res_trap}")
print(f"Сімпсон:  {res_simp}")

print("\n--- ТЕСТ 2: Графік ---")
solver.plot_convergence(0, 1, 1/3)
```
